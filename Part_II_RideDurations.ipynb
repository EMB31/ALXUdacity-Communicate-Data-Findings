## Investigation Overview
For this investigation I wanted to find out what factors affect ride durations. 


## Dataset Overview

The Ford GoBike System data set includes information about individual rides made in a bike-sharing system covering the greater San Francisco Bay area. Within this dataset, there is personal information about the users, duration of the ride, the start and end station amongst others. The datatypes for the Start_time, End_time and member_birth_year columns were changed.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df_FordGoBike = pd.read_csv('201902-fordgobike-tripdata.csv')
```

```python
#change the datatypes for the start_time, end_time and member_birth_year columns
#change NaN values in 'member_birth_year' column to 0
df_FordGoBike['member_birth_year'] = df_FordGoBike['member_birth_year'].fillna(0)

# convert member_birth_year from float to integer
df_FordGoBike = df_FordGoBike.astype({"member_birth_year":"int64"})

#convert start_time and end_time to datetime columns
df_FordGoBike['start_time'] = pd.to_datetime(df_FordGoBike['start_time'])
df_FordGoBike['end_time'] = pd.to_datetime(df_FordGoBike['end_time'])
```

# Distribution of Duration
Ride durations go as far as 10000 seconds. However majority of the ride durations are under 2000 seconds.

```python
bins=np.arange(60,10000,10)
plt.figure(figsize = [15, 8])
plt.hist(df_FordGoBike['duration_sec'], bins=bins);
plt.title("Distribution of ride duration",weight ='bold')
plt.xlabel("Ride Duration (seconds)", labelpad=15, weight='bold',size=12)
plt.ylabel("Frequency",labelpad=15, weight='bold',size=12);
```

## Gender distribution
There are 3 gender categories noted in the dataset; male, female and other. The males account for almost 80% of the total number of users, with females of about 23% and the rest of other gender.

```python
gender = df_FordGoBike['member_gender'].value_counts()
plt.figure(figsize = [5, 5])
plt.axis('square');
plt.pie(gender, labels = gender.index, wedgeprops = {'width': 0.7}, radius = 1.2, autopct ='%0.1f%%'); 
plt.title('Gender Distribution');
```

## Member birth years

With members born as early as the 1900s, and as late as the early 2000s; majority of the users are however born between the years 1980 to 2000.

```python
bins=np.arange(1875,df_FordGoBike['member_birth_year'].max()+5,1)
plt.figure(figsize = [8, 6]);
plt.hist(df_FordGoBike['member_birth_year'], bins=bins);
plt.title("Distribution of member birth years",weight ='bold')
plt.xlabel('Member Birth Years',labelpad=15, weight='bold',size=12);
plt.ylabel("Frequency",labelpad=15, weight='bold',size=12);
```

## Ride Duration against gender.

Though seemingly insignificant, males account for the longest ride duration, followed closely by Other gender then females.
There is also a greater saturation of female members in the low ride duration region.

```python
base_color = sb.color_palette()[0]
plt.figure(figsize = [10,12])
sb.violinplot(data=df_FordGoBike, x='member_gender', y='duration_sec', color=base_color);
plt.title("Ride Duration against Member Gender",weight ='bold')
plt.xlabel('Gender',labelpad=15, weight='bold',size=12);
plt.ylabel("Ride Duration in seconds",labelpad=15, weight='bold',size=12);
```

## Ride duration against member birth year

We notice that there are relatively more records for longer ride duration as the member birth year increases.

```python
plt.figure(figsize = [120, 20])
sb.boxplot(data=df_FordGoBike, x ='member_birth_year', y='duration_sec');
plt.title("Ride Duration against Birth Years",weight ='bold')
plt.xlabel('Member Birth Year',labelpad=15, weight='bold',size=20);
plt.ylabel("Ride Duration in seconds",labelpad=15, weight='bold',size=20);
```
